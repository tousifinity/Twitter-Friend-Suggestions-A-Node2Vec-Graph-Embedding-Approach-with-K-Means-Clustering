# Step 2: Data Preprocessing

### Convert Data Type

```python
import pandas as pd

# Assuming twitterData is a DataFrame

# Strip double quotes and attempt to convert 'id' and 'tweetId' columns to integers
twitterData['id'] = pd.to_numeric(twitterData['id'].str.strip('"'), errors='coerce')
twitterData['tweetId'] = pd.to_numeric(twitterData['tweetId'].str.strip('"'), errors='coerce')

# Strip double quotes from other columns
twitterData['screenName'] = twitterData['screenName'].str.strip('"')
twitterData['avatar'] = twitterData['avatar'].str.strip('"')
twitterData['lang'] = twitterData['lang'].str.strip('"')

# Display information about the DataFrame
twitterData.info()
```

### Kept Only Needed Column

```python
# Select and keep only the desired columns
twitterData = twitterData[['id', 'friends']]

# Now, twitterData contains only the 'id', and 'friends' columns
twitterData.head()
```

### Removing Empty Column

```python
# Check for null values in the DataFrame
null_check = twitterData.isnull().any(axis=1)

# If any column is empty in a row, remove that row
if null_check.any():
    twitterData = twitterData[~null_check]

# Now, twitterData does not contain rows where any column is empty
```

# Step 3: Model Training

```python
import networkx as nx
import pandas as pd
import ast  # Import the 'ast' module

# Load the updated dataset from your CSV file
updated_dataset = pd.read_csv('50_data.csv')

# Create an empty graph
G = nx.Graph()

# Iterate through your dataset and add edges to the graph
for index, row in updated_dataset.iterrows():
    user_id = str(row['id'])
    
    # Extract the friends from the 'friends' column
    friends = ast.literal_eval(row['friends'])
    
    for friend_id in friends:
        G.add_edge(user_id, friend_id)
```

### Node2Vec Implementation

```python
from node2vec import Node2Vec

# Create a Node2Vec object with your graph 'G'
# Precompute the probabilities and generate walks
node2vec = Node2Vec(G, dimensions=64, walk_length=30, num_walks=200, workers=4)  # Set the number of CPU cores

# Learn embeddings
model = node2vec.fit(window=10, min_count=1)

# Save the model to a file
model.save("node2vec_model_recommendation_50.gensim")
```

### K-means Clustering

```python
from gensim.models import Word2Vec
from sklearn.cluster import KMeans

# Load the Node2Vec model with embeddings
model = Word2Vec.load("node2vec_model_recommendation_50.gensim")

# Get the embeddings for all nodes
node_embeddings = {node: model.wv[node] for node in G.nodes}

# Specify the number of clusters (K) for K-means
num_clusters = 25  # You can adjust this number as needed

# Convert the node embeddings to a list
node_embeddings_list = list(node_embeddings.values())

# Initialize the K-means model
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Fit K-means to the node embeddings
kmeans.fit(node_embeddings_list)

# Get cluster assignments for each node
cluster_assignments = kmeans.labels_

# Print out the nodes in each cluster
clustered_nodes = {i: [] for i in range(num_clusters)}
for i, node in enumerate(G.nodes):
    cluster_id = cluster_assignments[i]
    clustered_nodes[cluster_id].append(node)

for cluster_id, nodes in clustered_nodes.items():
    cluster_str = ', '.join(nodes)
    print(f"\033[1mCluster {cluster_id}:\033[0m {cluster_str}")
    print()

```

### Graphical User Interface

```python
import PySimpleGUI as sg
import gensim

# Load the pre-trained Node2Vec model
loaded_model = gensim.models.Word2Vec.load("node2vec_model_recommendation_50.gensim")

# PySimpleGUI layout
sg.theme("Python")  # Set the theme to Dark Teal (you can choose other themes)
font_size = ("Helvetica", 20)  # You can adjust the font size here

layout = [
    [sg.Text("Enter a User ID: "), sg.InputText(key="user_id", size=(200, 2))],
    [sg.Button("Get Friend Suggestions", size=(30, 1)), sg.Button("Get Cluster", size=(30, 1)), sg.Button("Get All Clusters", size=(30, 1)), sg.Button("Exit", size=(30, 1))],
    [sg.Text("Friend Suggestions/Cluster Information:", size=(35, 1))],
    [sg.Listbox([], size=(200, 30), key="suggestions_clusters")]
]

# Create the window
window = sg.Window("Friend Suggestion & Cluster Info App", layout, size=(1000, 500))

# Function to make friend suggestions (similar nodes)
def make_friend_suggestions(user_id):
    try:
        suggestions = loaded_model.wv.most_similar([user_id], topn=20)
        return [str(suggestion[0]) for suggestion in suggestions]
    except KeyError:
        return ["User ID not found in the model."]

def get_cluster_info(user_id):
    try:
        user_index = list(G.nodes).index(user_id)  # Get the index of the user ID
        cluster_id = cluster_assignments[user_index]  # Get the cluster ID for the user
        users_in_cluster = [node for i, node in enumerate(G.nodes) if cluster_assignments[i] == cluster_id]
        return [
            f"User {user_id} is in Cluster {cluster_id}.",
            f"Users in this cluster: {', '.join(users_in_cluster)}"
        ]
    
    except ValueError:
        return ["User ID not found in the clusters."]
    
def get_all_clusters():
    cluster_results = {}
    for i, node in enumerate(G.nodes):
        cluster_id = cluster_assignments[i]
        if cluster_id not in cluster_results:
            cluster_results[cluster_id] = []
        cluster_results[cluster_id].append(node)
    
    # Sort the cluster_results by cluster ID
    sorted_cluster_results = sorted(cluster_results.items(), key=lambda item: item[0])
    
    cluster_info = [
        f"Cluster {cluster_id}: {', '.join(nodes)}"
        for cluster_id, nodes in sorted_cluster_results
    ]
    return cluster_info

# Event loop
while True:
    event, values = window.read()

    if event in (sg.WIN_CLOSED, "Exit"):
        break

    if event == "Get Friend Suggestions":
        user_id = values["user_id"]
        suggestions = make_friend_suggestions(user_id)
        window["suggestions_clusters"].update(suggestions)

    if event == "Get Cluster":
        user_id = values["user_id"]
        cluster_info = get_cluster_info(user_id)
        window["suggestions_clusters"].update(cluster_info)

    if event == "Get All Clusters":
        all_clusters = get_all_clusters()
        window["suggestions_clusters"].update(all_clusters)

# Close the window
window.close()
```

# Step 4: Accuracy

### Silhouette Score

```python
from sklearn.metrics import silhouette_score
import numpy as np

# Get cluster assignments for each node
cluster_assignments = kmeans.labels_

# Calculate Silhouette Score
# measures how similar an object is to its own cluster compared to other clusters
silhouette_avg = silhouette_score(node_embeddings_list, cluster_assignments, metric='euclidean')
print(f'Silhouette Score: {silhouette_avg:.5f}')
```

### ROC-AUC Score

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# Split the edges into training and test sets
edges = list(G.edges())
train_edges, test_edges = train_test_split(edges, test_size=0.2)

# Extract embeddings for nodes involved in edges
def get_node_embedding(node_id):
    return model.wv[node_id]

# Calculate ROC-AUC score for link prediction
y_true = [1] * len(test_edges)  # True links
y_true += [0] * len(train_edges)  # Non-links
edge_list = test_edges + train_edges
y_pred = [np.dot(get_node_embedding(u), get_node_embedding(v)) for u, v in edge_list]

roc_auc = roc_auc_score(y_true, y_pred)
print(f'ROC-AUC Score: {roc_auc:.5f}')
```

# Word2Vec

```python
import networkx as nx
import pandas as pd
import ast
from gensim.models import Word2Vec

# Load the dataset from your CSV file
dataset = pd.read_csv('50_data.csv')

# Create an empty graph
G = nx.Graph()

# Iterate through your dataset and add edges to the graph
for index, row in dataset.iterrows():
    user_id = str(row['id'])
    
    # Extract the friends from the 'friends' column
    friends = ast.literal_eval(row['friends'])
    
    for friend_id in friends:
        G.add_edge(user_id, friend_id)

# Extract the nodes from the graph
nodes_word2vec = list(G.nodes())

# Generate sentences for Word2Vec from the graph
sentences_word2vec = []
for node_word2vec in nodes_word2vec:
    friends_word2vec = [str(neighbor) for neighbor in G.neighbors(node_word2vec)]
    sentences_word2vec.append(friends_word2vec)

# Train Word2Vec model
model_word2vec = Word2Vec(sentences_word2vec, vector_size=64, window=10, min_count=1, sg=0, workers=4)

# Save the Word2Vec model to a file
model_word2vec.save("word2vec_model_recommendation_50.gensim")
```

### K-Means Clustering for Word2Vec

```python
from gensim.models import Word2Vec
from sklearn.cluster import KMeans

# Load the Word2Vec model
model_word2vec = Word2Vec.load("word2vec_model_recommendation_50.gensim")

# Get the embeddings for all nodes (users) in your model
user_embeddings = {user: model_word2vec.wv[user] for user in model_word2vec.wv.index_to_key}

# Specify the number of clusters (K) for K-means
num_clusters = 25  # You can adjust this number as needed

# Convert the user embeddings to a list
user_embeddings_list = list(user_embeddings.values())

# Initialize the K-means model
kmeans = KMeans(n_clusters=num_clusters, random_state=0)

# Fit K-means to the user embeddings
kmeans.fit(user_embeddings_list)

# Get cluster assignments for each user
cluster_assignments = kmeans.labels_

# Print out the users in each cluster
clustered_users = {i: [] for i in range(num_clusters)}
for i, user in enumerate(user_embeddings.keys()):
    cluster_id = cluster_assignments[i]
    clustered_users[cluster_id].append(user)

for cluster_id, users in clustered_users.items():
    user_str = ', '.join(users)
    print(f"\033[1mCluster {cluster_id}:\033[0m {user_str}")
    print()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
import numpy as np

# Split the edges into training and test sets
edges = list(G.edges())
train_edges, test_edges = train_test_split(edges, test_size=0.2, random_state=42)

# Extract embeddings for nodes involved in edges
def get_user_embedding(user_id):
    return model_word2vec.wv[user_id]

# Calculate ROC-AUC score for link prediction
y_true = [1] * len(test_edges)  # True links
y_true += [0] * len(train_edges)  # Non-links
edge_list = test_edges + train_edges
y_pred = [np.dot(get_user_embedding(u), get_user_embedding(v)) for u, v in edge_list]

roc_auc = roc_auc_score(y_true, y_pred)
print(f'ROC-AUC Score: {roc_auc:.5f}')
```
